```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
import numpy as np
```

```python
# ML Models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
```

```python
# pa_df.dropna(inplace=True)
```

```python
# Binning Continuous Addiction_Level → Categories
pa_df['Addiction_Level_Category'] = pd.cut(pa_df['Addiction_Level'],
    bins=[-np.inf, 3.5, 7.5, np.inf], # np.inf and -np.inf represent positive and negative infinity
    labels=[0, 1, 2]  # 0=Low, 1=Medium, 2=High
)
print(pa_df)
```

```python
# Replace original Addiction_Level with categorical version
pa_df.drop(columns=['Addiction_Level'], inplace=True)
pa_df.rename(columns={"Addiction_Level_Category": 'Addiction_Level'}, inplace=True)
# print(pa_df.columns)
```

```python
# Drop rare classes (<2 samples)
target_col = 'Addiction_Level'
class_counts = pa_df[target_col].value_counts()
valid_classes = class_counts[class_counts >= 2].index
df = pa_df[pa_df[target_col].isin(valid_classes)]
print(valid_classes)
```

```python
#Encode Categorical Columns
label_encoders = {}
for col in pa_df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    pa_df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
#Split Features and Target
X = pa_df.drop(target_col, axis=1)
y = pa_df[target_col].astype(int)  # ensure integer classes
```

```python
#Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# ML Models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(probability=True),
    "Gradient Boosting": GradientBoostingClassifier(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
    "Neural Net (MLP)": MLPClassifier(max_iter=500)
}
```

```python
#Train & Evaluate All Models
results = {}
for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {acc:.4f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    results[name] = acc
```

```python
# Accuracy Summary
results_df = pd.DataFrame(list(results.items()), columns=["Model", "Accuracy"]).sort_values(by="Accuracy", ascending=False)
print("\nModel Accuracy Summary:")
print(results_df)
```

```python
# Plot Accuracy
plt.figure(figsize=(10, 5))
sns.barplot(x="Accuracy", y="Model", data=results_df, palette='coolwarm')
plt.title("Model Accuracy Comparison")
plt.xlabel("Accuracy")
plt.ylabel("Model")
plt.xlim(0, 1)
plt.grid(True)
plt.show()
```
